```python
from glob import glob
import json
import pandas as pd

files = glob('./output/experiments_number_hidden_neurons/number_hidden_neurons_*.json')

all_dicts = []

for file in files:
    with open(file, "r") as f:
        data = json.load(f)
        all_dicts.append(data)

df = pd.DataFrame(all_dicts)
df = df.fillna(1)

df
```

```python
import matplotlib.pyplot as plt

x = df["number_hidden_neurons"]
test_acc = df["test_accuracy"]
time = df["time"]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))  # 1 row, 2 columns

axes[0].scatter(x, test_acc, marker='o', color='blue')
axes[0].set_title("Test Accuracy vs Number hidden layer")
axes[0].set_xlabel("Number hidden layer")
axes[0].set_ylabel("Test Accuracy")

axes[1].scatter(x, time, marker='^', color='green')
axes[1].set_title("Time vs Number hidden layer")
axes[1].set_xlabel("Number hidden layer")
axes[1].set_ylabel("Time (s)")

plt.tight_layout()
plt.show()
```
