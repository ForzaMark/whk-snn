```python
from util.create_data_loader import create_data_loader

train_data_loader, test_data_loader = create_data_loader(
    time_steps=100, use_train_subset=False, batch_size=1
)
```

```python
train_features = []
train_labels = []

for x,y in train_data_loader:
    features = x.squeeze()

    train_features.append(features.numpy())
    train_labels.append(y.item())    
    
```

```python
test_features = []
test_labels = []

for x,y in test_data_loader:
    features = x.squeeze()

    test_features.append(features.numpy())
    test_labels.append(y.item())  
```

```python
import numpy as np

np.save("../../e-prop-implementation/eligibility_propagation/datasets/hdd/full_spiking_data/hdd_test_features.npy", test_features)
np.save("../../e-prop-implementation/eligibility_propagation/datasets/hdd/full_spiking_data/hdd_train_features.npy", train_features)
np.save("../../e-prop-implementation/eligibility_propagation/datasets/hdd/full_spiking_data/hdd_test_labels.npy", test_labels)
np.save("../../e-prop-implementation/eligibility_propagation/datasets/hdd/full_spiking_data/hdd_train_labels.npy", train_labels)
```
